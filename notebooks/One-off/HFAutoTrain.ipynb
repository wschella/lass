```python
# https://huggingface.co/wschell/autotrain-lass-983132630
# https://huggingface.co/wschell/autotrain-lass-983132631
# https://huggingface.co/wschell/autotrain-lass-983132632
# https://huggingface.co/wschell/autotrain-lass-983132633
# https://huggingface.co/wschell/autotrain-lass-983132634
# https://huggingface.co/wschell/autotrain-lass-983132635
# https://huggingface.co/wschell/autotrain-lass-983132636
# https://huggingface.co/wschell/autotrain-lass-983132637
# https://huggingface.co/wschell/autotrain-lass-983132638
# https://huggingface.co/wschell/autotrain-lass-983132639
# https://huggingface.co/wschell/autotrain-lass-983132640
# https://huggingface.co/wschell/autotrain-lass-983132641
# https://huggingface.co/wschell/autotrain-lass-983132642
# https://huggingface.co/wschell/autotrain-lass-983132643
# https://huggingface.co/wschell/autotrain-lass-983132644

runs = [
    "983132630",
    "983132631",
    "983132632",
    "983132633",
    "983132634",
    "983132635",
    "983132636",
    "983132637",
    "983132638",
    "983132639",
    "983132640",
    "983132641",
    "983132642",
    "983132643",
    "983132644",
]
```

```python
import requests
from pathlib import Path

token = Path("../../.hftoken").read_text().strip()
```

```python
configs = []
readmes = []
for run in runs:
    config = requests.get(
        f"https://huggingface.co/wschell/autotrain-lass-{run}/raw/main/config.json",
        headers={"Authorization ": f"Bearer {token}"})
    configs.append(config.json())

    readme = requests.get(
        f"https://huggingface.co/wschell/autotrain-lass-{run}/raw/main/README.md",
        headers={"Authorization ": f"Bearer {token}"})
    readmes.append(readme.text)
```

```python
print(configs[0])
print(readmes[0])
```

```python
import pandas as pd
import re

aurocs = []
architecture = []
# learning_rate = []
max_length = []
hidden_act = []
dropout_hidden = []
dropout_attention = []
hidden_size = []
layer_norm_eps = []
vocab_size = []
position_embedding_type = []
sinusoidal_pos_embds = []
for conf, readme in zip(configs, readmes):
    auroc = float(re.findall(r"AUC: ([0-9.]+)", readme)[0])
    aurocs.append(auroc)

    architecture.append(conf["model_type"])
    print(conf["model_type"])
    # learning_rate.append(conf["learning_rate"])
    max_length.append(conf["max_length"])
    hidden_act.append(conf.get("hidden_act"))
    dropout_hidden.append(conf.get("hidden_dropout_prob"))
    dropout_attention.append(conf.get("attention_probs_dropout_prob"))
    hidden_size.append(conf.get("hidden_size"))
    layer_norm_eps.append(conf.get("layer_norm_eps"))
    vocab_size.append(conf.get("vocab_size"))
    position_embedding_type.append(conf.get("position_embedding_type"))
    sinusoidal_pos_embds.append(conf.get("sinusoidal_pos_embds"))


df = pd.DataFrame({
    "run": runs,
    "auroc": aurocs,
    "architecture": architecture,
    # "learning_rate": learning_rate,
    "max_length": max_length,
    "hidden_act": hidden_act,
    "dropout_hidden": dropout_hidden,
    "dropout_attention": dropout_attention,
    "hidden_size": hidden_size,
    "layer_norm_eps": layer_norm_eps,
    "vocab_size": vocab_size,
    "position_embedding_type": position_embedding_type,
    "sinusoidal_pos_embds": sinusoidal_pos_embds,
})
```

```python
df
```

```python
non_duplicates = df.drop(columns=["run", "auroc"]).drop_duplicates(keep="first").index

df_dedup = df.loc[non_duplicates]
df_dedup = df_dedup.sort_values("architecture", ascending=True)
df_dedup = df_dedup.drop(columns=["run", "dropout_attention", "dropout_hidden", "position_embedding_type", "hidden_act", "max_length"])
df_dedup
```
